```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
```

```python
# View first five rows of the dataset
df.head()
```

```python
df.dtypes
```

```python
counts = df['who'].value_counts()
# Plot the distribution of man, woman and child using a pie chart
sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("notebook", font_scale=1.2)
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
counts = df['who'].value_counts()
# Plot the distribution of man, woman and child using a pie chart
sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("notebook", font_scale=1.2)
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
# Countplot
sns.catplot(x ="who", hue ="survived",kind ="count", data = df)
```

```python
sns.histplot(df, x='fare', kde=True)
plt.title('Distribution of Ticket Fares')
plt.show()
```

Obeservations:
<br>1.There where 60.3% males, 30.4% females and 9.3% children on the ship. 
<br>2.We can observe that females and children survived in more number in this unfortunate event.
<br>3.Majority of people paid fare less than 50$

```python
# Plot the distribution of age for each gender, with information about survival
sns.boxplot(data=df, x='sex', y='age', hue='survived')
plt.title('Age Distribution by Gender and Survival')
plt.show()
```
