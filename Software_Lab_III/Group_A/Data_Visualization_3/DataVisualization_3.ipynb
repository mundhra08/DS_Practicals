```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

# Create a DataFrame from the data and target attributes
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target_names[iris.target]
```

```python
# Inference 1: List down the features and their types
# All features in the Iris dataset are numeric
print(df.dtypes)
```

```python
# Inference 2: Create a histogram for each feature
sns.histplot(data=df, x='sepal length (cm)', kde=True)
plt.title('Sepal Length Distribution')
plt.show()

sns.histplot(data=df, x='sepal width (cm)', kde=True)
plt.title('Sepal Width Distribution')
plt.show()

sns.histplot(data=df, x='petal length (cm)', kde=True)
plt.title('Petal Length Distribution')
plt.show()

sns.histplot(data=df, x='petal width (cm)', kde=True)
plt.title('Petal Width Distribution')
plt.show()
```

```python
# Inference 3: Create a boxplot for each feature
sns.boxplot(data=df, x='sepal length (cm)')
plt.title('Sepal Length Distribution')
plt.show()

sns.boxplot(data=df, x='sepal width (cm)')
plt.title('Sepal Width Distribution')
plt.show()

sns.boxplot(data=df, x='petal length (cm)')
plt.title('Petal Length Distribution')
plt.show()

sns.boxplot(data=df, x='petal width (cm)')
plt.title('Petal Width Distribution')
plt.show()
```

Inference 4: Compare distributions and identify outliers
<br>Based on the boxplots, there appear to be some outliers in the sepal width feature
