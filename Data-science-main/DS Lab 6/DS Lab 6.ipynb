```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv("iris.csv")
```

Following dataset has no null values hence this can be directly used to train the model
sepallength,sepalwidth and petallength,petalwidth are already scaled integral values, 

```python
x=df.iloc[:,1:5]
y=df.iloc[:,-1]
```

```python
sc=StandardScaler()
X=sc.fit_transform(x)
```

```python
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
nb=GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_pred=nb.predict(x_test)
```

```python
nb.score(x_test,y_test)*100
```

Hence our model has the accuracy of 96.67%

```python

#computing confusion metrics
cm= confusion_matrix(y_test,y_pred)

#Calculating TP,FN,TN,FP
TP = cm[1][1]
FN = cm[0][1]
TN = cm[0][0]
FP = cm[1][0]

print("TP ",TP)
print("FN ",FN)
print("TN ",TN)
print("FP ",FP)

#Calculating Accuracy, Error rate, Precision, Recall
accuracy = TP+TN/(TP+FN+TN+FP)
error_rate = FP+FN/(TP+FN+TN+FP)
precision = TP/(TP+FP)
recall = TP/(TP+FN)

print("Accuracy ",accuracy)
print("Error_rate ",error_rate)
print("Precision ",precision)
print("Recall ",recall)
```
